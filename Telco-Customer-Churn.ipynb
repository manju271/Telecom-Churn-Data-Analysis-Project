```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure SeniorCitizen is categorical
df["SeniorCitizen"] = df["SeniorCitizen"].map({0: "No", 1: "Yes"})

# Calculate counts and percentages
grouped = df.groupby(["SeniorCitizen", "Churn"]).size().unstack()
grouped = grouped.div(grouped.sum(axis=1), axis=0) * 100  # Convert to percentages

# Plot
plt.figure(figsize=(4,4))
ax = grouped.plot(kind="bar", stacked=True, colormap="coolwarm", figsize=(4,4), edgecolor="black")

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center", fontsize=10, color="white")

# Customizations
plt.title("Churn by Senior Citizen (Stacked %)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.xticks(rotation=0)
plt.legend(title='Churn',bbox_to_anchor = (0.9,0.9)) # Set y-axis to 100% scale
plt.show()
```

```python

# Define the columns for subplots
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set number of rows and columns for subplots
n_rows = 3  # Adjust based on the number of features
n_cols = 3  # 3x3 grid for 9 plots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 10))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Create count plots for each feature
for i, col in enumerate(columns):
    sns.countplot(x=df[col], ax=axes[i], palette="viridis", edgecolor="black",hue = df["Churn"])
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis="x", rotation=0)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

#The visualizations show the distribution of customer churn based on various telecom services such as Phone Service, Multiple Lines, Internet Service, 
Online Security, Online Backup, Device Protection, Tech Support, Streaming TV, andStreaming Movies. In general, churn is higher for customers who do 
not use these services or have limited access, particularly in categories like Fiber Optic Internet, Online Security, and Tech Support. 
The data suggests that providing these additional services may help reduce churn.
